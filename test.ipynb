### Toy model：兩團 Gaussian 在前向擴散下的解析 score

1. **資料分佈與 forward 過程**

- 初始資料分佈：
  $$
  x_0 \sim 0.5\,\mathcal{N}(\mu_1, I) + 0.5\,\mathcal{N}(\mu_2, I)
  $$
- 前向擴散（固定時間 $t$）：
  $$
  x_t = \sqrt{\bar\alpha_t}\,x_0 + \sigma_t \epsilon,\quad \epsilon \sim \mathcal{N}(0, I)
  $$

---

2. **條件在第 $k$ 團 Gaussian 下的 $x_t$ 分佈**

對於第 $k \in \{1,2\}$ 個 component，若
$$
x_0 \mid k \sim \mathcal{N}(\mu_k, I),
$$
則
$$
x_t \mid k \sim \mathcal{N}(m_{k,t}, \Sigma_t),
$$
其中
$$
m_{k,t} = \sqrt{\bar\alpha_t}\,\mu_k,
$$
$$
\Sigma_t = \bar\alpha_t I + \sigma_t^2 I = (\bar\alpha_t + \sigma_t^2)\,I.
$$

---

3. **時間 $t$ 的邊際分佈 $p_t(x_t)$**

因為是兩個 Gaussian 的 mixture，加上前向是線性高斯，所以
$$
p_t(x_t)
= 0.5\,\mathcal{N}(x_t; m_{1,t}, \Sigma_t)
+ 0.5\,\mathcal{N}(x_t; m_{2,t}, \Sigma_t).
$$

我們記
$$
q_k(x_t) := \mathcal{N}(x_t; m_{k,t}, \Sigma_t),\quad k=1,2,
$$
則
$$
p_t(x_t) = 0.5\,q_1(x_t) + 0.5\,q_2(x_t).
$$

---

4. **posterior 責任值 $\gamma_k(x_t, t)$**

定義第 $k$ 個 component 在觀察到 $x_t$ 下的 posterior 權重
$$
\gamma_k(x_t, t)
:= \mathbb{P}(k \mid x_t)
= \frac{0.5\,q_k(x_t)}{0.5\,q_1(x_t) + 0.5\,q_2(x_t)}
= \frac{q_k(x_t)}{q_1(x_t) + q_2(x_t)},
$$
因此
$$
\gamma_1(x_t, t) + \gamma_2(x_t, t) = 1.
$$

---

5. **單一 Gaussian 的 score**

若
$$
x_t \sim \mathcal{N}(m_{k,t}, \Sigma_t),
$$
則其 log-density 的梯度為
$$
\nabla_{x_t} \log q_k(x_t)
= -\Sigma_t^{-1}(x_t - m_{k,t}).
$$

---

6. **mixture 的真實 score：$s^\star(x_t, t) = \nabla_{x_t}\log p_t(x_t)$**

利用
$$
\nabla_{x_t} \log p_t(x_t)
= \sum_{k=1}^2 \gamma_k(x_t, t)\,\nabla_{x_t} \log q_k(x_t),
$$
得到
$$
s^\star(x_t, t)
:= \nabla_{x_t} \log p_t(x_t)
= -\sum_{k=1}^2 \gamma_k(x_t, t)\,\Sigma_t^{-1}(x_t - m_{k,t}).
$$

由於在 toy model 中
$$
\Sigma_t = (\bar\alpha_t + \sigma_t^2)\,I,
$$
因此
$$
\Sigma_t^{-1} = \frac{1}{\bar\alpha_t + \sigma_t^2}\,I,
$$
score 可寫成
$$
s^\star(x_t, t)
= -\frac{1}{\bar\alpha_t + \sigma_t^2}
\sum_{k=1}^2 \gamma_k(x_t, t)\,(x_t - m_{k,t}).
$$

---

7. **和 $-\epsilon/\sigma_t$ 的關係（期望形式）**

在 forward 過程中
$$
x_t = \sqrt{\bar\alpha_t}\,x_0 + \sigma_t \epsilon
\quad\Longrightarrow\quad
\epsilon = \frac{x_t - \sqrt{\bar\alpha_t}x_0}{\sigma_t}.
$$

真正的理論 score 可寫成條件期望：
$$
s^\star(x_t, t)
= \mathbb{E}\left[-\frac{\epsilon}{\sigma_t}\,\middle|\,x_t\right]
= -\frac{1}{\sigma_t^2}
\left(x_t - \sqrt{\bar\alpha_t}\,\mathbb{E}[x_0 \mid x_t]\right).
$$

在兩團 Gaussian 的情況下，
$$
\mathbb{E}[x_0 \mid x_t]
= \sum_{k=1}^2 \gamma_k(x_t, t)\,\mu_k,
$$
代回即可驗證這個形式與上面 mixture score 的寫法是一致的。
